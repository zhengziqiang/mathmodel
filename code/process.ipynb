```python
def angle(x1,y1,x2,y2):
    x1=torad(x1)
    y1=torad(y1)
    x2=torad(x2)
    y2=torad(y2)
    d=math.sin(y1)*math.sin(y2)+math.cos(y1)*math.cos(y2)*math.cos(x2-x1)
    d=np.sqrt(1.0-d*d)
    d=math.cos(y2)*math.sin(x2-x1)/d
    d=math.asin(d)
    if x1>x2:
        return math.atan((x2-x1)/(y2-y1))
    else:
        return math.pi+math.atan((x2-x1)/(y2-y1))
```

```python
def angle(x1,y1,x2,y2):
    x1=torad(x1)
    y1=torad(y1)
    x2=torad(x2)
    y2=torad(y2)
    d=math.sin(y1)*math.sin(y2)+math.cos(y1)*math.cos(y2)*math.cos(x2-x1)
    d=np.sqrt(1.0-d*d)
    d=math.cos(y2)*math.sin(x2-x1)/d
    d=math.asin(d)*180/math.pi
    return -d
```

```python
data=np.genfromtxt("C:\\Users\\Administrator\\Desktop\\数学建模\\b-1.csv",dtype=float,delimiter=',')
latitude=data[1:,1]
magtitude=data[1:,2]
```

```python
new_x=[]
new_y=[]
for i in range(len(latitude)):
    dis=transfer(magtitude[i],latitude[i],114.0,23.0)
    ang=angle(magtitude[i],latitude[i],114.0,23.0)
    ang=np.abs(ang)
    ang=torad(ang)
    if magtitude[i]>114.0 and latitude[i]>23.0:
        y=dis*math.cos(ang)
        x=dis*math.sin(ang)
    if magtitude[i]>114.0 and latitude[i]<23.0:
        y=-dis*math.cos(ang)
        x=dis*math.sin(ang)
    if magtitude[i]<114.0 and latitude[i]>23.0:
        y=dis*math.cos(ang)
        x=-dis*math.sin(ang)
    if magtitude[i]<114.0 and latitude[i]<23.0:
        y=-dis*math.cos(ang)
        x=-dis*math.sin(ang)
    new_x.append(x)
    new_y.append(y)
```

```python
import io
out=open("C:\\Users\\Administrator\\Desktop\\数学建模\\transfer-distance-1.csv","w")
for i in range(len(new_y)):
    out.write(str(new_x[i])+','+str(new_y[i])+'\n')
out.close()
```
